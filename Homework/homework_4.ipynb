### Условие 1: Задача 1
- 1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
import pandas as pd

df = pd.read_csv("kc_house_data.csv", sep=',', encoding='windows-1251')
df.head()
```

- 1.2. Изучите стоимости недвижимости

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.hist(df['price'])
plt.title('price')
plt.xlabel('price')
plt.ylabel('quantity')
plt.xticks(rotation=30)
# по гистограмме можно сделать вывод, что больше всего представлено недвижимости дешевой и чем дороже, 
# тем меньше по количеству вариантов
```

- 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(6,4))
plt.hist(df['sqft_living'])
plt.title('sqft_living')
plt.xlabel('sqft_living')
plt.ylabel('quantity')
plt.xticks(rotation=30)
# по гистограмме можно сделать вывод, что больше всего вариантов средней квадратуры, с маленькой на треть меньше чем средней, 
# а огромной квадратурой очень мало домов.
```

- 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(6,4))
plt.hist(df['yr_built'])
plt.title('yr_built')
plt.xlabel('yr_built')
plt.ylabel('quantity')
plt.xticks(rotation=30)
# по гистограмме видно, что больше всего домов примерно с 2010 - 2015 года, 
# по количеству на треть меньше домов примерно с 1950 - 2010 года,
# и меньше всего старых домов с 1900 - 1950 года
```

### Условие 3: 3 задача
- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.head()
df.sort_values('price')
```

```python
data = df.groupby('bedrooms')['price'].mean().reset_index()
data
plt.figure(figsize=(6,4))
plt.scatter(data['bedrooms'], data['price'])
plt.title('price')
plt.xlabel('bedrooms')
plt.ylabel('price')

# по точечному графику можно сделать вывод, что рост стоимости неджимости идет от 1 до 6 комнат вверх, 
# где 6 комнат там максимальная стоимость за недвижимость, а цена падает вниз
```

```python
data = df.groupby('floors')['price'].agg(mean='mean').reset_index().sort_values(by='floors')
data
plt.figure(figsize=(6,4))
plt.plot(data['floors'], data['mean'])
plt.title('price')
plt.xlabel('floors')
plt.ylabel('price')
# по линейному графику мы видим, что стоимость недвижимости выше всего у 2,5 этажных домов, далее 3.5 этажные.
# самые дешевые 1 этажные дома
```

```python

data = df.groupby('yr_built')['price'].mean().reset_index()
data

plt.figure(figsize=(6,4))
plt.bar(data['yr_built'],data['price'])
plt.title('price')
plt.xlabel('yr_built')
plt.ylabel('price')
plt.xticks(rotation=30)

# по диаграмме видно, что цена в зависимости от года сильно не зависит, так как в любые года есть как и высокие цены, так и низкие
```

```python
data = pd.crosstab(index=df['view'], columns=df['waterfront'], values=df['price'], aggfunc='mean').reset_index()
data.head()
```

```python
import numpy as np
```

```python
plt.figure(figsize=(10,4))
n_ticks = np.arange(len(data['view']))
offset = 0.2
w = 0.4
plt.bar(n_ticks - offset, data[0], width=w)
plt.bar(n_ticks + offset, data[1], width=w)
plt.title('tabl')
plt.xlabel('view')
plt.ylabel('price')
plt.legend(['no', 'yes'])
plt.xticks(n_ticks, data['view'], rotation= 30)
# по диаграмме видно, что средняя цена очень зависима от качества вида и есть ли вид на набережную, 
# мы видим, что при виде 2 и когда есть вид на набережную - самая высокая цена, далее идет вид 4 и есть вид на набережную
```
